 to buy a new lip-stick is shown in table below. (Implement step by step
 using commands- Dont use library) Use this dataset to build a decision
 tree, with Buys as the target variable, to help in buying lipsticks in the
 future. Find the root node of the decision tree.

```python
# Program 4
# Build decision tree root node manually (entropy & information gain)

import math
import pandas as pd

# Example structure  replace rows with your actual dataset
data = [
    # {"Age": "<21", "Income": "High", "Gender": "Female", "MaritalStatus": "Single", "Buys": "Yes"},
    # ...
]

df = pd.DataFrame(data)

def entropy(labels):
    probs = labels.value_counts(normalize=True)
    return -sum(p * math.log2(p) for p in probs)

def info_gain(df, attribute, target):
    total_entropy = entropy(df[target])
    values = df[attribute].unique()
    weighted_entropy = 0.0
    for v in values:
        subset = df[df[attribute] == v]
        weight = len(subset) / len(df)
        weighted_entropy += weight * entropy(subset[target])
    return total_entropy - weighted_entropy

target_col = "Buys"
features = ["Age", "Income", "Gender", "MaritalStatus"]

print("Total entropy:", entropy(df[target_col]))

gains = {}
for feat in features:
    gains[feat] = info_gain(df, feat, target_col)
    print(f"Information gain for {feat}: {gains[feat]}")

root_node = max(gains, key=gains.get)
print("\nRoot node of the decision tree is:", root_node)
```

 points. P1=[0.1,0.6] P2=[0.15,0.71] P3=[0.08,0.9] P4=[0.16, 0.85]
 P5=[0.2,0.3] P6=[0.25,0.5] P7=[0.24,0.1] P8=[0.3,0.2]. Perform the k-mean
 clustering with initial centroids as m1=P1 =Cluster#1=C1 and
 m2=P8=cluster#2=C2. Answer the following 1] Which cluster does P6
 belong to? 2] What is the population of a cluster around m2? 3] What is
 the updated value of m1 and m2?

```python
# # Program 9

import numpy as np

points = np.array([
    [0.1, 0.6],   # P1
    [0.15, 0.71], # P2
    [0.08, 0.9],  # P3
    [0.16, 0.85], # P4
    [0.2, 0.3],   # P5
    [0.25, 0.5],  # P6
    [0.24, 0.1],  # P7
    [0.3, 0.2]    # P8
])

# initial centroids: m1 = P1, m2 = P8
m1 = points[0].copy()
m2 = points[7].copy()

def distance_formula(a, b):
    return np.sqrt(np.sum((a - b) ** 2))

def assign_clusters(points, c1, c2):
    clusters = {1: [], 2: []}
    for i, p in enumerate(points):
        d1 = distance_formula(p, c1)
        d2 = distance_formula(p, c2)
        if d1 < d2:
            clusters[1].append(i)
        else:
            clusters[2].append(i)
    return clusters

def recompute_centroid(points, indices):
    if len(indices) == 0:
        return None
    return np.mean(points[indices], axis=0)

clusters = assign_clusters(points, m1, m2)
print("Initial clusters:", clusters)

m1_new = recompute_centroid(points, clusters[1])
m2_new = recompute_centroid(points, clusters[2])

print("Updated m1:", m1_new)
print("Updated m2:", m2_new)

# Answers:
print("1] Which cluster does P6 belong to?")
for c, idxs in clusters.items():
    if 5 in idxs:  # P6 index = 5
        print("P6 belongs to cluster:", c)

print("2] Population around m2 (cluster 2):", len(clusters[2]))
print("3] Updated m1 and m2:", m1_new, m2_new)

# ----



import math

# Points
P = [(0.1,0.6),(0.15,0.71),(0.08,0.9),(0.16,0.85),
     (0.2,0.3),(0.25,0.5),(0.24,0.1),(0.3,0.2)]

# Initial centroids
m1 = P[0]   # P1
m2 = P[7]   # P8

# Assign each point to nearest cluster
C1, C2 = [], []       
for p in P:
    d1 = math.dist(p, m1)
    d2 = math.dist(p, m2)
    (C1 if d1 < d2 else C2).append(p)

# Update centroids
m1_new = (sum(x for x,y in C1)/len(C1),  sum(y for x,y in C1)/len(C1))
m2_new = (sum(x for x,y in C2)/len(C2),  sum(y for x,y in C2)/len(C2))

print("Cluster 1:", C1)
print("Cluster 2:", C2)
print("\nUpdated m1:", m1_new)
print("Updated m2:", m2_new)

# Answer: Where is P6?
print("\nP6 belongs to:", "Cluster 1" if P[5] in C1 else "Cluster 2")
print("Population of cluster around m2:", len(C2))
```

10. Write a program to do the following: You have given a collection of 8
 points. P1=[2, 10] P2=[2, 5] P3=[8, 4] P4=[5, 8] P5=[7,5] P6=[6, 4] P7=[1, 2]
 P8=[4, 9]. Perform the k-mean clustering with initial centroids as m1=P1
 =Cluster#1=C1 and m2=P4=cluster#2=C2, m3=P7 =Cluster#3=C3. Answer
 the following 1] Which cluster does P6 belong to? 2] What is the
 population of a cluster around m3? 3] What is the updated value of m1,
 m2, m3?


```python
# Program 10

import numpy as np

points = np.array([
    [2, 10], # P1
    [2, 5],  # P2
    [8, 4],  # P3
    [5, 8],  # P4
    [7, 5],  # P5
    [6, 4],  # P6
    [1, 2],  # P7
    [4, 9]   # P8
])

# m1 = P1, m2 = P4, m3 = P7
m1 = points[0].copy()
m2 = points[3].copy()
m3 = points[6].copy()

def distance_formula(a, b):
    return np.sqrt(np.sum((a - b) ** 2))

def assign_clusters3(points, c1, c2, c3):
    clusters = {1: [], 2: [], 3: []}
    for i, p in enumerate(points):
        d1 = distance_formula(p, c1)
        d2 = distance_formula(p, c2)
        d3 = distance_formula(p, c3)
        dmin = min(d1, d2, d3)
        if dmin == d1:
            clusters[1].append(i)
        elif dmin == d2:
            clusters[2].append(i)
        else:
            clusters[3].append(i)
    return clusters

def recompute(points, indices):
    if not indices:
        return None
    return np.mean(points[indices], axis=0)

clusters = assign_clusters3(points, m1, m2, m3)
print("Initial clusters:", clusters)

m1_new = recompute(points, clusters[1])
m2_new = recompute(points, clusters[2])
m3_new = recompute(points, clusters[3])

print("Updated centroids:")
print("m1:", m1_new)
print("m2:", m2_new)
print("m3:", m3_new)

# Answers
print("1] Cluster of P6:")
for c, idxs in clusters.items():
    if 5 in idxs:  # P6 index = 5
        print("P6 belongs to cluster:", c)

print("2] Population around m3 (cluster 3):", len(clusters[3]))
print("3] Updated m1, m2, m3:", m1_new, m2_new, m3_new)
```

```python
# Program 23

import math
import pandas as pd

# Data from the image
data = [
    {"Age": "Young",  "Income": "High",   "Married": "No",  "Health": "Fair", "Class": "No"},
    {"Age": "young",  "Income": "High",   "Married": "No",  "Health": "Good", "Class": "No"},
    {"Age": "Middle", "Income": "High",   "Married": "No",  "Health": "Fair", "Class": "Yes"},
    {"Age": "Old",    "Income": "Medium", "Married": "No",  "Health": "Fair", "Class": "Yes"},
    {"Age": "Old",    "Income": "Low",    "Married": "Yes", "Health": "Fair", "Class": "Yes"},
    {"Age": "Old",    "Income": "Low",    "Married": "Yes", "Health": "Good", "Class": "No"},
    {"Age": "Middle", "Income": "Low",    "Married": "Yes", "Health": "Good", "Class": "Yes"},
    {"Age": "Young",  "Income": "Medium", "Married": "No",  "Health": "Fair", "Class": "No"},
    {"Age": "Young",  "Income": "Low",    "Married": "Yes", "Health": "Fair", "Class": "Yes"},
    {"Age": "Old",    "Income": "Medium", "Married": "Yes", "Health": "Fair", "Class": "Yes"},
    {"Age": "Young",  "Income": "Medium", "Married": "Yes", "Health": "Good", "Class": "Yes"},
    {"Age": "Middle", "Income": "Medium", "Married": "No",  "Health": "Good", "Class": "Yes"},
    {"Age": "Middle", "Income": "High",   "Married": "Yes", "Health": "Fair", "Class": "Yes"},
    {"Age": "Old",    "Income": "Medium", "Married": "No",  "Health": "Good", "Class": "No"},
]

df = pd.DataFrame(data)

# Frequency table for Age vs Class
freq_table = pd.crosstab(df["Age"], df["Class"])
print("Frequency Table (Age vs Class):")
print(freq_table)

def entropy(labels):
    probs = labels.value_counts(normalize=True)
    return -sum(p * math.log2(p) for p in probs)

def info_gain(df, attribute, target):
    total_entropy = entropy(df[target])
    values = df[attribute].unique()
    weighted_entropy = 0.0
    for v in values:
        subset = df[df[attribute] == v]
        weight = len(subset) / len(df)
        weighted_entropy += weight * entropy(subset[target])
    return total_entropy - weighted_entropy

target_col = "Class"
attribute = "Age"
gain_age = info_gain(df, attribute, target_col)

print("\nInformation Gain when splitting on Age:", gain_age)
```

25. Perform Data Cleaning, Data transformation using Python on any data set.

```python
# Program 25

import pandas as pd

df = pd.read_csv("your_dataset.csv")  # replace

print("Original shape:", df.shape)

# 1. Remove duplicate rows
df = df.drop_duplicates()
print("After dropping duplicates:", df.shape)

# 2. Handle missing values
for col in df.columns:
    if df[col].isna().sum() > 0:
        if pd.api.types.is_numeric_dtype(df[col]):
            df[col].fillna(df[col].median(), inplace=True)
        else:
            df[col].fillna("Unknown", inplace=True)

# 3. Remove outliers using IQR for a numeric column as example
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower) & (df[col] <= upper)]

print("After removing outliers:", df.shape)

# 4. Data transformation  normalization / standardization
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_scaled = df.copy()
df_scaled[num_cols] = scaler.fit_transform(df[num_cols])

print("Final cleaned & transformed data:")
print(df_scaled.head())
```
